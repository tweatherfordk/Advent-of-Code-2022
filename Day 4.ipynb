```python
with open('AOC Data/AOC_Day4_input.txt', 'r') as f:
    lines = f.readlines()
    data = [entry.strip() for entry in lines]
    
data
```

# Part 1
- Does one range contain another?

```python
pairs = 0

for entry in data:
    first, second = entry.split(",")
    
    first = [int(i) for i in first.split("-")]
    second = [int(i) for i in second.split("-")]
    
    if first[0] <= second[0] and first[1] >= second[1]:
        pairs +=1
    elif second[0] <= first[0] and second[1] >= first[1]:
        pairs +=1
        
print("Answer for part 1: ", pairs)
```

# Part 2
- Range overlap?

```python
pairs = 0

for entry in data:
    first, second = entry.split(",")
    
    first = [int(i) for i in first.split("-")]
    second = [int(i) for i in second.split("-")]
    
    if first[0] is range(second[0], second[1]+1) or first[1] in range(second[0], second[1]+1):
        pairs += 1
    elif second[0] is range(first[0], first[1]+1) or second[1] in range(first[0], first[1]+1):
        pairs += 1

print("Answer for part 2: ", pairs)
```
